## 1.4 Solución del problema

```python
import numpy as np
import pandas as pd
import pandas_datareader as web
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
closes=web.DataReader(name="SBUX",data_source="yahoo",start="2018-10-31",end=None)
starbucks=closes["Adj Close"]
closes=web.DataReader(name="NKE",data_source="yahoo",start="2018-10-31",end=None)
nike=closes["Adj Close"]
closes=web.DataReader(name="SAIC",data_source="yahoo",start="2018-10-31",end=None)
saic=closes["Adj Close"]
closes=web.DataReader(name="PAGS",data_source="yahoo",start="2018-10-31",end=None)
pags=closes["Adj Close"]

stocks=pd.concat([starbucks,nike,saic,pags],axis=1)
stocks.columns=["Starbucks","Nike","SAIC","PAGS"]
stocks.plot(figsize=(10,8));
plt.title("Stocks Portfolio");
plt.xlabel("Date");
plt.ylabel("USD");
```

```python
log_ret=np.log(stocks/stocks.shift(1))
log_ret.head()
```

```python
log_ret.hist(bins=100,figsize=(12,8))
plt.tight_layout()
```
